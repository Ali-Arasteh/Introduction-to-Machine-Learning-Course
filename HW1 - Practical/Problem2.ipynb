## Loading dataset
Load the IMDB-Movie-Data.csv file into a pandas dataframe and show its first 5 rows. Each row of this dataframe corresponds to one film.

```python
data = pd.read_csv("IMDB-Movie-Data.csv")
pd.set_option('display.max_rows', data.shape[0] + 1)
data.head(5)
```

## Calculating Profit
Add a column to the dataframe that calculate the profit of the movies. (Suppose the inflation rate is 3% and 1990 as the base year)

```python
data.insert(11, 'Profit (Based on 1990)', data['Revenue (Millions)'] * (1.03 ** (1990 - data['Year'])))
data.head(10)
```

## Nolan Movies
Filter Christopher Nolan movies and sort them in a descending order with respect to their profits then show the result.

```python
CNM = data[data['Director'] == 'Christopher Nolan'].sort_values(by='Profit (Based on 1990)', ascending=False)
CNM
```

## Favorite Genres
Calculate the revenues of different genres in a dataframe and sort them in a descending order. Then show the result.

```python
# First, we remove all movies whose 'Revenue' value is NaN
temp = data[[not math.isnan(i) for i in data['Revenue (Millions)'].values.tolist()]]
temp = temp['Genre'].str.get_dummies(sep=',')
Genre = temp.sum().index.values.tolist()
C = temp.sum()
S = temp.mul(data['Revenue (Millions)'], axis=0).sum()
RM = (S / C).tolist()
SGR = pd.DataFrame({'Genre': Genre, 'Revenue Mean': RM}).sort_values(by='Revenue Mean', ascending=False)
SGR.head(10)
```

## Actors/Actresses Activity
For each Actors/Actresses caculate the number of movies in the dataframe, then sort the actors/actresses first by their number of movies in descending order and then by their name in ascending order. 

```python
temp = data['Actors'].str.get_dummies(sep=',').sum()
Actor = [i.strip() for i in temp.index.values.tolist()]
Count = temp.to_numpy()
A = pd.DataFrame({'Actor': Actor, 'Count': Count}).groupby(['Actor']).sum()
BC = A.sort_values(by='Count', ascending=False)
BC.head(10)
```

```python
BA = A.sort_values(by='Actor')
BA.head(10)
```

## Correlations
Calculate the correlation between rating, votes and metascore of the movies.
Is there any meaningful relation between these features? 

```python
Corr = data[['Rating', 'Votes', 'Metascore']].corr()
Corr
```
