## ECDF and Functions 

Write a function that takes a one-dimensional NumPy array as input and returns two arrays of x and y which together are going to draw an empirical distribution function or ECDF. 

```python
def ecdf(data):
    N = len(data)
    x, count = np.unique(data, return_counts=True)
    y = np.cumsum(count) / N
    return x, y
```

Plot the ECDF of petal length for all three species in a single window.

```python
x0, y0 = ecdf(data[target == 0, 2])
x1, y1 = ecdf(data[target == 1, 2])
x2, y2 = ecdf(data[target == 2, 2])
plt.figure(0, figsize=(8, 6.4))
plt.plot(x0, y0, 'b')
plt.plot(x1, y1, 'g')
plt.plot(x2, y2, 'r')
plt.title('ECDF of petal length for all three species')
plt.xlabel('Petal Length')
plt.ylabel('ECDF')
plt.grid()
```

## Swarm Plots

First, import pandas and seaborn libraries. Then convert the sklearn dataset you had to a data frame. Draw four [swarm plot](https://seaborn.pydata.org/generated/seaborn.swarmplot.html) that should look like this:

![](img3.png)

The Y-axis represents one of your features each time, and the columns in the X-axis represent different species of iris.

```python
tar = np.zeros(len(target)).astype('U')
tar[target == 0] = 'Setosa'
tar[target == 1] = 'Versicolour'
tar[target == 2] = 'Virginica'
df = pd.DataFrame({'SL': data[:, 0].tolist(), 'SW': data[:, 1].tolist(), 'PL': data[:, 2].tolist(), 'PW': data[:, 3].tolist(), 'target': tar.tolist()})
plt.figure(0, figsize=(8, 6.4))
sns.swarmplot(x='target', y='SL', data=df)
title = plt.title('Swarm Plot of SL')
plt.figure(1, figsize=(8, 6.4))
sns.swarmplot(x='target', y='SW', data=df)
title = plt.title('Swarm Plot of SW')
plt.figure(2, figsize=(8, 6.4))
sns.swarmplot(x='target', y='PL', data=df)
title = plt.title('Swarm Plot of PL')
plt.figure(3, figsize=(8, 6.4))
sns.swarmplot(x='target', y='PW', data=df)
title = plt.title('Swarm Plot of PW')
```
